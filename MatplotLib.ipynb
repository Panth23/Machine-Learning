Useful for making Plots

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

Plotting the data

```python
# sin value
plt.plot(x,y)
plt.show()
```

```python
# adding title x-asix and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2 # x square
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+') # red color and with plus sign
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  # 0,0 is the origin and 1,1 is the width and height
languages = ["Hindi", "Spanish", "English", "French", "Italian"]
people = [100,20,200,34,13]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('PEOPLE')
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ["Hindi", "Spanish", "English", "French", "Italian"]
people = [100,20,200,34,13]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
